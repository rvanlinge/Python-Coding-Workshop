```python
'''
Creating a list of values of projectile motion
'''
import matplotlib.pyplot as plt

v0 = 200.0
time = np.linspace(0,50,100)
g = 9.81

x = []
y = []

for i in time:
    x.append(v0*i*np.cos(80*np.pi/180))
    y.append(-0.5*g*i**2+v0*i*np.sin(80*np.pi/180))

x = np.array(x)
y = np.array(y)

plt.figure(figsize=(10,8))
plt.scatter(x,y,color='k',marker='o')
plt.xlabel('Horizontal Position')
plt.ylabel('Vertical Position')
plt.minorticks_on()
plt.show()
```

```python
def vertProjectile(time,v0,y0=0,theta=0,g=9.81):
    '''
    Equation of vartical motion for a projectile.
    Variables:
    t = time
    v0 = inital velocity
    y0 = initial starting point (default = 0)
    theta = angle of projectile being fired; must be given in degrees (default = 0)
    g = gravity (default = 9.81)
    '''
    t = time
    theta = (theta*np.pi)/180 # This is converting degrees to radians
    y = -0.5*g*t**2+v0*t*np.sin(theta)+y0
    
    return np.round(y)
    
```

```python
time = np.linspace(0,10,100)
v0 = 100

position = vertProjectile(v0=v0,time=time,theta=45)

plt.figure(figsize=(10,8))
plt.plot(time,position,color='k')
plt.show()
```

```python
def plotting(x,y):
    plt.figure(figsize=(10,8))
    plt.plot(x,y,color='k')
    plt.minorticks_on()
    plt.show()
    
    return None 
```

```python
def horiProjectile(t,v0,x0=0,theta=0,g=9.81):
    '''
    The general equation of horizontal motion for a projectile
    Variables:
    t = time
    v0 = inital velocity
    x0 = initial starting point (default = 0)
    theta = angle of projectile being fired (default = 0)
    g = gravity (default = 9.81)
    '''
    theta = (theta*np.pi)/180
    x = v0*t*np.cos(theta)+x0
    return np.round(x)
```

```python
time = np.linspace(0,10,50)
v0 = 50
theta = 43

y = vertProjectile(time,v0,theta = theta)
x = horiProjectile(time,v0,theta = theta)
plotting(x,y)
```

```python
time = np.linspace(0,10,50)
v0 = 50

position1 = vertProjectile(time,v0)
position2 = vertProjectile(time,v0,y0=50,theta=42)

plotting(time,position1)
plotting(time,position2)
```
