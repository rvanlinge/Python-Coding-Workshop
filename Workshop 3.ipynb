## Introduction to Numpy 

Before we get into functions, I want to give an overview of the library __Numpy__. It is a powerful and fundamental library that is used for scientific computations.  Their math functions are (general) much faster computationally, where it becomes very apparent in complex calculations. It also can do quickly do linear algebra, fourier analysis, and is used in most of other major scientific libraries. But one of the biggest places __Numpy__ shines are in its arrays.


Numpy arrays are powerful objects that lets one do calculations, manipulate, and reshape versus a Python list which is very limited on what we can (practically) use it for. Let's look at a few examples:

```python
'''
Let's define a Python list of random numbers and try doing a calculations to the whole list
'''

x = [10,5,6,9,80,70,2,.3,56,8]

# We can't do it
#print(x+5)

# We can only do it to individual indices 
x[4]+5

# This isn't what we want!!!
print(x*5) 
```

Now let's look at doing the same calculations on an array

```python
import numpy as np

xx = np.array([10,5,6,9,80,70,2,.3,56,8])

print(xx+5)

print(xx*5)
```

Also a feature of numpy arrays is the ability to find things such as mean, max, standard deviation, and minimum without having to calculate it yourself.

```python
print(xx.mean(),xx.min(),xx.max(),xx.std())
```

```python
'''
We can also create two arrays and do calculations. The one thing is that their dimensions must be the same!
'''

yy = np.random.randint(0,100,10)

print(yy)
```

```python
print(xx+yy)
```

```python
print(yy**xx)
```

```python
'''
We can check certain things between the two arrays
'''

print(xx<yy)
print(xx in yy)
```

```python
'''
We can restructure arrays depending on conditions. This will compare index to index (e.g will compare xx[0] to yy[0])
'''

print(xx in yy)

print(yy[yy<xx])
```

There are many ways to create an array: np.arange(),np.linspace(),np.blank(). Let's take a look at the different ways:

```python
'''
1) np.arange(min,max,stepsize) 
    This will NOT include the max value
    Does not handle floating step sizes
'''

x = np.arange(0,100) # first is a starting point, second is ending point, also a third option for the step size
print(x)

x = np.arange(0,100,10)
print(x)

```

```python
'''
2) np.linspace(min,max,stepsize)
    DOES include the max value
    Better for float step sizes
'''

x = np.linspace(0,20)
print(x)

x = np.linspace(0,1,200)
print(x)
```

```python
'''
3) Other ways
'''

x = np.array(range(10)) 
print(x)

x = np.array([x for x in range(100) if x % 2 == 0]) # gets all the even values from 0-100
print(x)
```

While powerful, there is one downside - it is difficult to create a blank array when you don't know how many values you'll use. It isn't as straightforward as list where we create a list and just append to the list. Numpy arrays are stored in memory and if you append to an array, it creates a new array in memory. You can also create an array of zeros but again you would have to know how many values you want to use.

My trick for this is to start out with a python list, do all the calculations/storing of values and when you're done covert the python list to a numpy array.

```python
'''
Creating a list of values of projectile motion
'''
import matplotlib.pyplot as plt

v0 = 200.0
time = np.linspace(0,50,100)
g = 9.81

x = []
y = []

for i in time:
    x.append(v0*i*np.cos(80*np.pi/180))
    y.append(-0.5*g*i**2+v0*i*np.sin(80*np.pi/180))

x = np.array(x)
y = np.array(y)

plt.figure(figsize=(10,8))
plt.scatter(x,y,color='k',marker='o')
plt.xlabel('Horizontal Position')
plt.ylabel('Vertical Position')
plt.minorticks_on()
plt.show()
```

## Functions

So why do we care about functions? One reason is it can make our code look a lot cleaner. If we're using a specific function over and over again, it's easier to define a function and use it's name versus having to type it out all the time. Also, we can remove nested functions  Another reason is that if this is a fundamental function used in physics, you can start your own library of functions and have them readily available! Again, let's go back to our projectile motion equation and see how we can simplify this code. 

The one function we were using often were the different components (x and y) of the motion. This would be a good reason to define a function:

y(t) = -0.5(g)(t)^2+v0(t)(sin(theta))+y0

x(t) = v0(t)(cos(theta))+x0

```python
def vertProjectile(t,v0,y0=0,theta=0,g=9.81):
    '''
    Equation of vertical motion for a projectile.
    Variables:
    t = time
    v0 = inital velocity
    y0 = initial starting point (default = 0)
    theta = angle of projectile being fired (default = 0)
    g = gravity (default = 9.81)
    '''
    
    
    theta = (theta*np.pi)/180 # converting to radians
    y = -0.5*g*t**2+v0*t*np.sin(theta)+y0
    return np.round(y)
```

```python
vertProjectile(5,10)
```

```python
time = np.linspace(0,10,50)
v0 = 50

position = vertProjectile(time,v0)
```

```python
plt.figure(figsize=(10,8))
plt.plot(time,position,color='k',linestyle='-')
plt.show()
```

```python
time = np.linspace(0,10,50)
v0 = 50

position1 = vertProjectile(time,v0)
position2 = vertProjectile(time,v0,y0=50,theta=42)
```

```python
plt.figure(figsize=(10,8))
plt.plot(time,position1,color='k',linestyle='-')
plt.plot(time,position2,color='darkred',linestyle='-.')
plt.show()
```

```python
'''
Have them do this
'''

def horiProjectile(t,v0,x0=0,theta=0,g=9.81):
    '''
    The general equation of horizontal motion for a projectile
    Variables:
    t = time
    v0 = inital velocity
    x0 = initial starting point (default = 0)
    theta = angle of projectile being fired (default = 0)
    g = gravity (default = 9.81)
    '''
    theta = (theta*np.pi)/180
    x = v0*t*np.cos(theta)+x0
    return np.round(x)
```

```python
time = np.linspace(0,10,50)
v0 = 50
theta = 43

y = vertProjectile(time,v0,theta = theta)
x = horiProjectile(time,v0,theta = theta)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x,y,color='k',linestyle='-')
plt.xlabel('Horizontal Position')
plt.ylabel('Vertical Position')
plt.minorticks_on()
plt.show()
```

```python
time = np.linspace(0,10,1000)
v1 = 70
theta1 = 37.5

v2 = 65
y0_2 = 100
theta2 = -30

x1 = horiProjectile(time,v1,theta=theta1)
y1 = vertProjectile(time,v1,theta=theta1)

x2 = horiProjectile(time,v2,theta=theta2)
y2 = vertProjectile(time,v2,theta=theta2,y0=y0_2)

plt.figure(figsize=(10,8))
plt.plot(x1,y1,color='darkgreen',label='Projectile 1')
plt.plot(x2,y2,color='darkred',label='Projectile 2')
plt.legend(loc='best')
plt.show()

plt.figure(figsize=(10,8))
plt.title('X position')
plt.plot(time,x1,color='green',label='Projectile 1')
plt.plot(time,x2,color='red',label='Projectile 2')
plt.legend(loc='best')
plt.show()

plt.figure(figsize=(10,8))
plt.title('Y Position')
plt.plot(time,y1,color='green',label='Projectile 1')
plt.plot(time,y2,color='red',label='Projectile 2')
plt.legend(loc='best')
plt.show()
```

So did they collide?

There's a bit of ambiguity here. Graphically we see that their paths cross, but the question is do they cross at the same time?
We can build a for-loop to solve this.


```python
'''
Comparing the two arrays
'''

x_match = []
y_match = []

for i in range(len(y1)):
    if x2[i]-1 <= x1[i] <= x2[i]+1:
        if y2[i]-2 <= y1[i] <= y2[i]+2:
            x_match.append(x1[i])
            y_match.append(y1[i])
        else:
            continue
    else:
        continue

if len(x_match) == 0:
    print('They did not collide')
else:
    x_match = np.array(x_match)
    y_match = np.array(y_match)
    
    plt.figure(figsize=(10,8))
    plt.plot(x1,y1,color='darkgreen',label='Projectile 1')
    plt.plot(x2,y2,color='darkred',label='Projectile 2')
    plt.scatter(x_match,y_match,color='darkorange',marker='x')
    plt.legend(loc='best')
    plt.show()
        
        


```
